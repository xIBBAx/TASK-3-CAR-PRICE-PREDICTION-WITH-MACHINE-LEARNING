### Creating a function to change the amounts to numerical Values

```python
def change_price_to_numeric(r):
    return float(("".join(r.split(" ")[0])).replace(",",""))
```

```python
df['Average price'] = df['Average price'].astype(str).apply(change_price_to_numeric)
df['Minimum price'] = df['Minimum price'].astype(str).apply(change_price_to_numeric)
df['Maximum price'] = df['Maximum price'].astype(str).apply(change_price_to_numeric)
```

```python
df1.info()
```

```python
df1.drop_duplicates(inplace=True)

df1.reset_index(drop=True)
```

```python
def transmission_type(d):
    lis_= ["Manual","M/T"," MT","manual"]
    if any(x in d for x in lis_):
        return "Manual"
    else:
        return "Automtic"
```

```python
df1['Transmission_type'] = df1['Car Model'].apply(transmission_type)
df1['Transmission_type'].value_counts()
```

```python
df1["Car Brand"] = df1['Car Model'].apply(lambda x:"".join(x.split()[0]))
df1["Car_Model_Year"] = df1['Car Model'].apply(lambda x:"".join(x.split()[-1]))
```

```python
df1[['Car Brand',"Car_Model_Year"]]
```

```python
df1['Car_Brand_Description'] = df1['Car Model'].apply(lambda x:" ".join(x.split()[1:-1]))

df1['Car_Brand_Description'].head(5).to_list()
```

```python
def car_model_description(car):
    lis = ['Automatic',"Automotic","Automtic","A/T/","A/T","A/T ","M/T","manual","Manual‏","manual‏","A/T","/","M/T/","Manual","AT-6DCT","AT","اتوماتيك"]
    dict_ = {"High line":"Highline","Base Line":"Baseline","High Line":"Highline","Top Line":"Topline"}
    for i in car:
        lis_1= car.split(" ")
        lis_2= [i for i in lis_1 if i not in lis]   
        for i in range(len(lis_2)):
            if lis_2[i] in dict_:
                lis_2[i] = dict_[lis_2[i]]
            lis_3= " ".join(lis_2)
            return lis_3
```

```python
df1['Car_Brand_Description'] = df1['Car_Brand_Description'].apply(car_model_description)
```

```python
df1['Car_Brand_Type'] = df1['Car_Brand_Description'].apply(lambda r:r.split(" ")[0])
```

```python
df1['Car_Brand_Type'] = df1["Car Brand"] + " " + df1['Car_Brand_Type']
```

```python
df1['Car_Brand_Type'].value_counts().to_frame().reset_index().sort_values(by="Car_Brand_Type",ascending=False).head(10)
```

```python
df1[["Old Price","Price Change","New Price"]].head(5)
```

```python
df1['Old Price']= df1['Old Price'].apply(lambda r:r[:-4].replace(',',"")).astype("float")

df1['New Price']= df1['New Price'].apply(lambda r:r[:-4].replace(',',"")).astype("float")

df1['Price Change'] = df1['New Price'] - df1['Old Price']
```

```python
df1['Price Change']
```

```python
df1['date_range']=pd.to_datetime(df1['date_range'],format="%d/%m/%Y",errors="coerce")

df1['Month'] = df1['date_range'].dt.month_name()
df1['Month_Num'] = df1['date_range'].dt.month
df1['Quarter'] = df1['date_range'].dt.quarter.astype(str).apply(lambda r:r.split(".")[0])
df1['Year'] = df1['date_range'].dt.year.astype(str).apply(lambda r:r.split(".")[0])
```
